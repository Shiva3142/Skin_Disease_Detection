```python
classes_list=['Acne and Rosacea Photos',
 'Atopic Dermatitis Photos',
 'Bullous Disease Photos',
 'Cellulitis Impetigo and other Bacterial Infections',
 'Herpes HPV and other STDs Photos',
 'Light Diseases and Disorders of Pigmentation',
 'Lupus and other Connective Tissue diseases',
 'Melanoma Skin Cancer Nevi and Moles',
 'Nail Fungus and other Nail Disease',
 'Poison Ivy Photos and other Contact Dermatitis',
 'Psoriasis pictures Lichen Planus and related diseases',
 'Scabies Lyme Disease and other Infestations and Bites',
 'Seborrheic Keratoses and other Benign Tumors',
 'Urticaria Hives',
 'Vasculitis Photos']
def printClassList():
    for class_name in classes_list:
        print(f"{classes_list.index(class_name)} -->  {class_name}")
printClassList()
def getClassvalue(index):
    print(classes_list[index])
```

## Model Definition

```python
model1_using_only_one_layers=keras.models.Sequential(
     [
         keras.layers.Conv2D(filters=32,kernel_size=3,strides=(1,1),padding='valid',activation='relu',input_shape=[224,224,1]),
         keras.layers.MaxPool2D(pool_size=(2,2)),
         keras.layers.Flatten(),
         keras.layers.Dense(units=512,activation='relu'),
         keras.layers.Dense(units=15,activation='softmax')
     ]
 )
```

```python
model1_using_only_one_layers
```

```python
model1_using_only_one_layers.summary()
```

```python
model1_using_only_one_layers.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model1_using_only_one_layers.fit(Training_Features,numpy.array(Training_Labels),epochs=1,verbose=1,validation_data=(Validation_Features,numpy.array(Validation_Labels)))
```

```python
model1_using_only_one_layers.save('model1_using_only_one_layers_trial_Data.h5')
```

```python
trained_model=tensorflow.keras.models.load_model('model1_using_only_one_layers_trial_Data.h5')
```

```python
prediction=trained_model.predict(numpy.expand_dims(Training_Features[2],axis=0)).round(5)*100
```
